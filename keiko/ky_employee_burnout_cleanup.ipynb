```python
import pandas as pd
```

```python
# Name of Csv file
test_file = '../Resources/test.csv'
train_file = '../Resources/train.csv'
```

```python
test_df = pd.read_csv(test_file, encoding="ISO-8859-1")
train_df = pd.read_csv(train_file, encoding="ISO-8859-1")
```

```python
test_df.head()
```

```python
#drop Employee ID 
test_df.drop('Employee ID', axis='columns', inplace=True)

test_df
```

```python
#Drop nulls
test_df.dropna()
```

```python
test_df.dtypes
```

```python
# convert data to binary 

test_df_clean = {'Gender': {'Male': 0, 'Female':1},
                 'Company Type': {'Service': 0, 'Product': 1},
                 'WFH Setup Available' : {'No': 0, 'Yes': 1}}
```

```python
# create new df with binary columns
test_new_df = test_df.replace(test_df_clean)

test_new_df.head()
```

```python
# rename columns for easier use later
test_new_df.rename(columns = {'Date of Joining' : "date_of_joining", 'Gender': 'gender', 'Company Type': 'company_type',
                         'WFH Setup Available':'WFH_setup_available', 'Designation':'designation', 
                         'Resource Allocation':'resource_allocarion', 'Mental Fatigue Score':'mental_fatigue_score'})
```

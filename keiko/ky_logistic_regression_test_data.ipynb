```python
import pandas as pd
```

```python
test_df = pd.read_csv("data/test_new.csv")
# Drop the null columns where all values are null
# df = df.dropna(axis='columns', how='all')
# Drop the null rows
test_df = test_df.dropna()
test_df = test_df.drop(columns = ['Date of Joining'])

test_df.count()
```

```python
test_df
```

```python
selected_features = test_df.drop(columns = ['Gender'])

selected_features
```

```python
#split train test
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(selected_features, test_df['Gender'], random_state=1)
```

```python
X_train
```

```python
#Scale data
from sklearn.preprocessing import MinMaxScaler

X_scaler = MinMaxScaler().fit(X_train)
X_scale_train = X_scaler.transform(X_train)
X_scale_test = X_scaler.transform(X_test)
```

```python
#Train the model 
from sklearn.linear_model import LogisticRegression

classifier = LogisticRegression()
classifier
```

```python
classifier.fit(X_train, y_train)
```

```python
print(f"Training Data Score: {classifier.score(X_train, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test, y_test)}")
```

```python
predictions = classifier.predict(X_test)
print(f"First 10 Predictions:   {predictions[:10]}")
print(f"First 10 Actual labels: {y_test[:10].tolist()}")
```

```python
pd.DataFrame({"Prediction": predictions, "Actual": y_test}).reset_index(drop=True)
```

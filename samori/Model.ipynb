```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
X = traindf.Resource_Allocation.values.reshape(-1, 1)
y = traindf.Burn_Rate.values.reshape(-1, 1)

print("Shape: ", X.shape, y.shape)

X
```

```python
plt.scatter(X, y)
plt.title("Linear Regression Resource Allocation vs Mental Fatigue")
plt.xlabel("Resource Allocation")
plt.ylabel("Burnout Rate")
plt.savefig("Lin_Reg_Burn_v_Resource.png")
plt.show()
```
